### plotshow

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

### 

```python
plt.plot(x,y1,color='green')
plt.plot(x,y2,color='purple')
plt.show()
```

### (xlim,ylim)

```python
plt.plot(x,y1,color='green',linestyle=':')
plt.plot(x,y2,color='purple')

plt.xlim(-2,10)
plt.ylim(-2,2)

plt.show()
```

#### axis

```python
plt.plot(x,y1,color='green',linestyle=':')
plt.plot(x,y2,color='purple')
plt.axis([-5,15,-2,2])
plt.show()
```

### xlabel/ylabel

```python
plt.plot(x,y1,color='green',linestyle=':')
plt.plot(x,y2,color='purple')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

### plotlabelplt.legend,plt.title

```python
plt.plot(x,y1,color='green',linestyle=':',label='sin(x)')
plt.plot(x,y2,color='purple',label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.title('First plot')
plt.show()
```
