### 显示多个图像调用多次plot再show

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

### 改变线条颜色

```python
plt.plot(x,y1,color='green')
plt.plot(x,y2,color='purple')
plt.show()
```

### 改变曲线形状，以及调整坐标范围(xlim,ylim)

```python
plt.plot(x,y1,color='green',linestyle=':')
plt.plot(x,y2,color='purple')

plt.xlim(-2,10)
plt.ylim(-2,2)

plt.show()
```

#### 通过axis方法一次性设置坐标范围传入列表

```python
plt.plot(x,y1,color='green',linestyle=':')
plt.plot(x,y2,color='purple')
plt.axis([-5,15,-2,2])
plt.show()
```

### 通过xlabel/ylabel设置坐标轴名字

```python
plt.plot(x,y1,color='green',linestyle=':')
plt.plot(x,y2,color='purple')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

### 在plot里用参数label为曲线设置名字并调用plt.legend显示名字,并通过plt.title为图表设置名字

```python
plt.plot(x,y1,color='green',linestyle=':',label='sin(x)')
plt.plot(x,y2,color='purple',label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.title('First plot')
plt.show()
```

## 数据是离散的不调用plot用scatter
## plot用于一个特征的情况x是特征y是观测值，scatter用于xy都是特征的情况通过散点图可以看它们的分布情况

```python
plt.scatter(x,y1,color='green')
plt.scatter(x,y2)
plt.show()
```

```python
x=np.random.normal(0,1,10000)
y=np.random.normal(0,1,10000)
plt.scatter(x,y)
plt.show()
```

#### 通过参数alpha设置散点图透明度，0完全透明，1完全不透明，取值范围0到1

```python
x=np.random.normal(0,1,10000)
y=np.random.normal(0,1,10000)
plt.scatter(x,y,alpha=0.1)
plt.show()
```
