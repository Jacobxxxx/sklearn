# 数据集

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w,b=1.8,2.5
```

```python
np.random.seed(0)
x1=np.random.rand(100)*4
noise=np.random.randn(100)
x2=w*x1+b+noise
```

```python
x1.shape,x2.shape
```

```python
x=np.vstack([x1,x2]).T
```

```python
x.shape
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

## 零均值化

```python
np.mean(x,axis=0)
```

```python
x-=np.mean(x,axis=0)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
np.mean(x,axis=0)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=1)
pca.fit(x)
```

```python
pca.components_
```

```python
np.array([pca.components_[0][0]*-1,pca.components_[0][0]]),
```

```python
np.array([pca.components_[0][1]*-1,pca.components_[0][1]])
```

```python
plt.scatter(x[:,0],x[:,1])
plt.plot(
    np.array([pca.components_[0][0]*-1,pca.components_[0][0]])*5,
    np.array([pca.components_[0][1]*-1,pca.components_[0][1]])*5,
    c='r'
)
plt.show()
```

```python
x_pca=pca.transform(x)
```

```python
x_pca.shape
```

```python
plt.scatter(x_pca,np.zeros_like(x_pca),s=10)
plt.show()
```

## 通过矩阵点乘将主成分反向投影回原始空间

```python
x_pca_inv=pca.inverse_transform(x_pca)
```

```python
plt.scatter(x[:,0],x[:,1],s=10)
plt.scatter(x_pca_inv[:,0],x_pca_inv[:,1],s=10,c='r')
plt.show()
```
