# 构造非线性数据

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.uniform(-4,2,size=100)
y=2*x**2+4*x+3+np.random.randn(100)
X=x.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.show()
```

# 线性回归

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X,y)
y_predict=lin_reg.predict(X)
```

```python
plt.plot(x,y_predict,c='r')
plt.scatter(x,y)
plt.show()
```

# 多项式回归

```python
X[:5]
```

```python
X_new=np.hstack([X,X**2])
X_new[:5]
```

```python
new_lin_reg=LinearRegression()
new_lin_reg.fit(X_new,y)
y_predict_new=new_lin_reg.predict(X_new)
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict_new[np.argsort(x)],c='r')
plt.show()
```

```python
new_lin_reg.intercept_
```

```python
new_lin_reg.coef_
```

# sicikit-learn中的PolynomialFeatures类为原样本添加新特征操作进行了封装

## [PolynomialFeatures](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html#sklearn.preprocessing.PolynomialFeatures)

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_features=PolynomialFeatures(degree=2)
```

```python
X_poly=polynomial_features.fit_transform(X)
```

```python
X_poly[:5]
```

```python
new_lin_reg=LinearRegression()
new_lin_reg.fit(X_poly,y)
y_predict_new=new_lin_reg.predict(X_poly)
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict_new[np.argsort(x)],c='r')
plt.show()
```

```python
new_lin_reg.intercept_
```

```python
new_lin_reg.coef_
```
