# 加载数据集

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
x=data[:,-1].reshape(-1,1)
y=target.reshape(-1,1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
plt.scatter(x_train,y_train,s=10)
plt.show()
```

# sklearn实现一元线性回归

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
model=lin_reg.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,c='r')
plt.show()
```

# MSE

```python
y_real =y_test
```

```python
mse=np.sum((y_real-y_predict)**2)/len(y_test)
mse
```

## from sklearn.metrics import mean_squared_error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_real,y_predict)
```

# RMSE

```python
rmse =np.sqrt(mse)
rmse
```

# MAE

```python
mae =np.sum(np.abs(y_real-y_predict)/len(y_test))
mae
```

## from sklearn.metrics import mean_absolute_error

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_real,y_predict)
```

# $R^2$

```python
r2=1-(np.sum((y_real-y_predict)**2))/(np.sum((y_real-np.mean(y_real))**2))
r2
```

```python
from sklearn.metrics import r2_score
r2_score(y_real,y_predict)
```
