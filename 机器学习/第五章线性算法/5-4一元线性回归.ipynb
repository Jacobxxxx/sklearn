# 一元线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
x=data[:,5]
y=target
x=x[y<50]#在y=50过多点要排除
y=y[y<50]
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=233)
plt.scatter(x_train,y_train)
plt.show()
```

# 一元线性回归公式实现

```python
def fit(x,y):
    a_up=np.sum((x-np.mean(x))*(y-np.mean(y)))
    a_bottow=np.sum((x-np.mean(x))**2)
    a=a_up/a_bottow
    b=np.mean(y)-a*np.mean(x)
    return a,b
```

```python
a,b=fit(x,y)
a,b
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,a*x_train+b,c='r')
plt.show()
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,a*x_test+b,c='r')
plt.show()
```

# sklearn 实现一元线性回归


```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train.reshape(-1,1),y_train)
```

## coef为数组对应k，intercept为数对应b

```python
y_predict=lin_reg.predict(x_test.reshape(-1,1))
a = lin_reg.coef_[0]
b = lin_reg.intercept_
print("a =", lin_reg.coef_)
print("b =", lin_reg.intercept_)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,c='r')
plt.show()
```
